```python
import numpy as np                       # Operaciones numéricas
import pandas as pd                      # Manejo de datos en tablas
import matplotlib.pyplot as plt          # Gráficas
from scipy import interpolate            # Funciones de interpolación
from sklearn.model_selection import train_test_split  # Separar datos
from sklearn.metrics import mean_squared_error, mean_absolute_error  # Métricas de error
from sklearn.preprocessing import MinMaxScaler        # Normalización de datos
from scipy.interpolate import CubicSpline, PchipInterpolator, BarycentricInterpolator
```

***Datos***

***5. Interpolaciones (Lagrange, Cubic Spline, PCHIP)***

```python
df = data_pandas.copy()


x = df['rho[g/cm^3]'].values
y = df['P[dyn/cm^2]'].values
order = np.argsort(x)
x, y = x[order], y[order]


x_fine = np.linspace(x.min(), x.max(), 1500) #secuencia

# Nodos para Lagrange (evita extremos)
NUM_NODOS = 10
idx_nodes = np.linspace(0, len(x)-1, NUM_NODOS+2, dtype=int)[1:-1]
x_nodes, y_nodes = x[idx_nodes], y[idx_nodes]

# Interpoladores
lag_bary = interpolate.BarycentricInterpolator(x_nodes, y_nodes)
cs_nat = interpolate.CubicSpline(x, y, bc_type='natural')
pchip = interpolate.PchipInterpolator(x, y)

# Evaluación
mask_interior = (x_fine >= x_nodes.min()) & (x_fine <= x_nodes.max())
y_lagrange_fine = np.full_like(x_fine, np.nan)
y_lagrange_fine[mask_interior] = lag_bary(x_fine[mask_interior])
y_cs_fine = cs_nat(x_fine)
y_pchip_fine = pchip(x_fine)

# Gráficas global + zoom
plt.figure(figsize=(13,5))

# Global
plt.subplot(1,2,1)
plt.plot(x, y, 'k.', ms=4, label='Datos')
plt.plot(x_fine, y_cs_fine, label='Cubic Spline')
plt.plot(x_fine, y_pchip_fine, label='PCHIP')
plt.plot(x_fine, y_lagrange_fine, color='orange', label='Lagrange')
plt.plot(x_nodes, y_nodes, 's', color='orange', ms=6, label='Nodos Lagrange')
plt.title("Interpolaciones (global)"); plt.xlabel("ρ [g/cm³]"); plt.ylabel("P [dyn/cm²]")
plt.legend(); plt.grid(True)

# Zoom en intervalo interior
plt.subplot(1,2,2)
plt.plot(x, y, 'k.', ms=4, label='Datos')
plt.plot(x_fine, y_cs_fine, label='Cubic Spline')
plt.plot(x_fine, y_pchip_fine, label='PCHIP')
plt.plot(x_fine[mask_interior], y_lagrange_fine[mask_interior], color='orange', label='Lagrange')
plt.plot(x_nodes, y_nodes, 's', color='orange', ms=6)
plt.title("Zoom interior Lagrange"); plt.xlabel("ρ [g/cm³]"); plt.ylabel("P [dyn/cm²]")
plt.legend(); plt.grid(True)

plt.tight_layout(); plt.show()
```

**6. División train / test + métricas**

```python
# Variables + normalización log(P)
x = data_pandas['rho[g/cm^3]'].values.reshape(-1,1)
y = data_pandas['P[dyn/cm^2]'].values.reshape(-1,1)
use_logP = True
y_proc = np.log10(y) if use_logP else y

scaler_x, scaler_y = MinMaxScaler(), MinMaxScaler()
x_s = scaler_x.fit_transform(x); y_s = scaler_y.fit_transform(y_proc)

# Train / Test split
x_train, x_test, y_train, y_test = train_test_split(x_s, y_s, test_size=0.2, random_state=42)

# Reconstrucción inversa de escala
def inv_transform_y(y_norm_arr):
    y_log_or_raw = scaler_y.inverse_transform(np.array(y_norm_arr).reshape(-1,1))
    return 10**(np.clip(y_log_or_raw.ravel(), -300, 300)) if use_logP else y_log_or_raw.ravel()

# Interpoladores con datos train
x_train_orig = scaler_x.inverse_transform(x_train).ravel()
y_train_orig = scaler_y.inverse_transform(y_train).ravel()
order = np.argsort(x_train_orig)
x_train_sorted, y_train_sorted = x_train_orig[order], y_train_orig[order]

cs = CubicSpline(x_train_sorted, y_train_sorted, bc_type='natural')
pchip = PchipInterpolator(x_train_sorted, y_train_sorted)
lagrange_interp = BarycentricInterpolator(x_train_sorted[:10], y_train_sorted[:10])  # ejemplo con 10 nodos

# Evaluación test
x_test_orig = scaler_x.inverse_transform(x_test).ravel()
y_test_orig = inv_transform_y(y_test)
y_cs_test = inv_transform_y(cs(x_test_orig))
y_pchip_test = inv_transform_y(pchip(x_test_orig))
y_lag_test = inv_transform_y(lagrange_interp(x_test_orig))

# Métricas
def metrics(y_true, y_pred):
    mask = np.isfinite(y_true) & np.isfinite(y_pred)
    return mean_squared_error(y_true[mask], y_pred[mask]), mean_absolute_error(y_true[mask], y_pred[mask])

print("=== Métricas en log(P) ===")
print(f"Cubic Spline: MSE={metrics(y_test.ravel(), scaler_y.transform(cs(x_test_orig).reshape(-1,1)).ravel())[0]:.3e}")
print(f"PCHIP       : MSE={metrics(y_test.ravel(), scaler_y.transform(pchip(x_test_orig).reshape(-1,1)).ravel())[0]:.3e}")
print(f"Lagrange    : MSE={metrics(y_test.ravel(), scaler_y.transform(lagrange_interp(x_test_orig).reshape(-1,1)).ravel())[0]:.3e}")

print("\n=== Métricas en escala original ===")
print(f"Cubic Spline: MSE={metrics(y_test_orig, y_cs_test)[0]:.3e}")
print(f"PCHIP       : MSE={metrics(y_test_orig, y_pchip_test)[0]:.3e}")
print(f"Lagrange    : MSE={metrics(y_test_orig, y_lag_test)[0]:.3e}")
```
